```python
import pandas as pd
import numpy as np
```

# Read all the files 

```python
df1 = pd.read_csv("C:/Users/melvi/Desktop/Capstone/final_1.csv",encoding = "gbk")
```

```python
df_geo = pd.read_excel("C:/Users/melvi/Desktop/Capstone/geocode.xlsx",encoding = "gbk")
```

```python
df_dist = pd.read_csv("C:/Users/melvi/Desktop/Capstone/distance.csv",encoding = "gbk")
```

# merge files

```python
df_final = pd.merge(df1, df_dist, on = "Community")
```

```python
df_final
```

# Create Dummies

```python
df_decoration = pd.get_dummies(df_final["Decoration"])
df_decoration.head()
```

```python
df_floor = pd.get_dummies(df_final["Floor"])
df_floor.head()
```

```python
df_orientation = pd.get_dummies(df_final["Orientation"])
df_orientation.head()
```

```python
df_neigh = pd.get_dummies(df_final["Neighborhood"])
df_neigh.head()
```

```python
df_district = pd.get_dummies(df_final["District"])
df_district.head()
```

```python
df_final[["室","厅"]] = df_final["Layout"].str.extract("(\d+)室(\d+)厅")
```

```python
df_final.head()
```

# Concat all the dummies to the dataset

```python
df = pd.concat([df_final,df_decoration,df_orientation,df_floor,df_neigh,df_district],axis=1)
```
